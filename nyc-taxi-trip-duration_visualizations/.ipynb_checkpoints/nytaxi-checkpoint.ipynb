```python
#NY_Taxi dataset
# Reference: https://www.kaggle.com/gaborfodor/from-eda-to-the-top-lb-0-367/comments
%matplotlib inline
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from datetime import timedelta
import datetime as dt
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [32, 20]
import seaborn as sns
#import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.cluster import MiniBatchKMeans
from sklearn.preprocessing import StandardScaler
```

```python
#read data once
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')
```

```python
#dataframe.shape[0] or dataframe.shape[1] gives the length along the y and x direction respectively.
print ("train_df has %s rows and %s coloumns" % (train_df.shape[0],train_df.shape[1]))
print ("train_df has %s rows and %s coloumns" % (test_df.shape[0],test_df.shape[1]))
train_df.head()
```

## Handling the time and date data
Let's covert the string date and time in train and test files into datetime datatypes in python.

```python

train_df['pickup_datetime'] = pd.to_datetime(train_df.pickup_datetime)
train_df['dropoff_datetime'] = pd.to_datetime(train_df.dropoff_datetime)

test_df['pickup_datetime'] = pd.to_datetime(test_df.pickup_datetime)

#separating pickup date and pickup time
train_df.loc[:,'pickup_date'] = train_df['pickup_datetime'].dt.date
test_df.loc[:,'pickup_date'] = test_df['pickup_datetime'].dt.date

train_df.loc[:,'pickup_time'] = train_df['pickup_datetime'].dt.time
test_df.loc[:,'pickup_time'] = test_df['pickup_datetime'].dt.time

#Converting 'Y' in 'store_and_forward' column to Boolean 1 and 0.
train_df['store_and_fwd_flag'] = 1 * (train_df.store_and_fwd_flag.values == 'Y')
test_df['store_and_fwd_flag'] = 1 * (test_df.store_and_fwd_flag.values == 'Y')

#output shows that the string values in pickup_datetime have been converted to timestamps.
```

## Check trip durations
We should check whether trip durations in the original training data are correct or not. We calculate
difference between given duration with the one we will calculate.

```python

train_df['calculated_duration'] = (train_df['dropoff_datetime'] - train_df['pickup_datetime']).map(lambda x: x.total_seconds())
#Check !!
if (np.array_equal(train_df['calculated_duration'], train_df['trip_duration'])): 
    print ('All Iz Well !!') 
else: print("Durations not equal !!")
```

# Principal Component Analysis
The data can be represented in terms of its principal components. This is useful in dimensionality reduction and aides the tree classifiers. In this case we do not intend to do dimensionality reduction but rather a high variance representation of the data to aide the tree classifiers. First we standardize the data so that all features have a mean of 0 and unit variance.

```python
#Data Standardization
coords_train = np.vstack((train_df[['pickup_latitude', 'pickup_longitude']].values,
                    train_df[['dropoff_latitude', 'dropoff_longitude']].values))


scaler = StandardScaler().fit(coords_train) #fit StandardScaler to training data.

pca_train = PCA().fit(scaler.transform(coords_train)) #fit StandardScaler to training data.

# pca_test = PCA(n_components = 2).fit(coords_test)
# print (pca_train.explained_variance_ratio_)
# print (pca_test.explained_variance_ratio_)
# train = (pca_train.components_)
#test = (pca_test.components_)

#Transforming the coordinates of the train and test data and writing them as new features in the dataframe
train_df['pickup_pca_0'] = pca_train.transform(scaler.transform(train_df[['pickup_latitude', 'pickup_longitude']]))[:,0]
train_df['pickup_pca_1'] = pca_train.transform(scaler.transform(train_df[['pickup_latitude', 'pickup_longitude']]))[:,1]
train_df['dropoff_pca_0'] = pca_train.transform(scaler.transform(train_df[['dropoff_latitude', 'dropoff_longitude']]))[:,0]
train_df['dropoff_pca_1'] = pca_train.transform(scaler.transform(train_df[['dropoff_latitude', 'dropoff_longitude']]))[:,1]

test_df['pickup_pca_0'] = pca_train.transform(scaler.transform(test_df[['pickup_latitude', 'pickup_longitude']]))[:,0]
test_df['pickup_pca_1'] = pca_train.transform(scaler.transform(test_df[['pickup_latitude', 'pickup_longitude']]))[:,1]
test_df['dropoff_pca_0'] = pca_train.transform(scaler.transform(test_df[['dropoff_latitude', 'dropoff_longitude']]))[:,0]
test_df['dropoff_pca_1'] = pca_train.transform(scaler.transform(test_df[['dropoff_latitude', 'dropoff_longitude']]))[:,1]

plt.scatter(train_df['pickup_pca_1'],train_df['pickup_pca_0'],s=.1, label='train', alpha=0.08)
plt.ylim(-5,4)
figsize=(10,10)
plt.xlim(-5,4)
plt.title('Principal Components of Standardized pickup locations')
plt.xlabel('pickup_pca_1')
plt.ylabel('pickup_pca_0')
plt.show()
```

## Distances between pickup and dropoff points
The coordinates of pickup and dropoff points for each trip can be used to calculate a rough estimate of distance travelled by a taxi during a trip. However there are a few caveats:
- The haversine distance between two points is a poor estimate of distance travelled as taxis only travel on intersecting streets. Haversine distance is the shortest distance between two points on the surface of a sphere.
- Manhattan distance is a better estimate of distance travelled around city blocks. But is it an accurate estimate of the distance? It assumes that city streets are aligned in north-south and east-west directions. This is not the case for New York. Reference : http://www.charlespetzold.com/etc/AvenuesOfManhattan/
- After a rough calculation on a scratch paper, I realized that the orientation of the axis of New York does not affect the "Manhattan Distance". So I am going to use it as the distance travelled by the taxi.

```python
#let's define a fuction manhattan() that calculates manhattan distance between two cooridnates.
def manhattan(lat_pickup,long_pickup,lat_dropoff,long_dropoff):
    dist = abs(lat_pickup-lat_dropoff) + abs(long_pickup-long_dropoff)
    return dist

train_df['distance_manhattan'] = manhattan(train_df['pickup_latitude'],train_df['pickup_longitude'],train_df['dropoff_latitude'],train_df['dropoff_longitude'])
test_df['distance_manhattan'] = manhattan(test_df['pickup_latitude'],test_df['pickup_longitude'],test_df['dropoff_latitude'],test_df['dropoff_longitude'])
```
